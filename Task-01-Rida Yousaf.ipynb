```python
# my chatbot project for decodelabs
# project 1 - rule based chatbot

# dictionary to store all responses
responses = {
    "hello": "Hey! How can i help you?",
    "hi": "Hi there! What do you want to know?",
    "hey": "Hey! Ask me something.",
    "bye": "Bye! See you later.",
    "goodbye": "Goodbye! Take care.",
    "how are you": "I am doing great thanks for asking!",
    "what are you": "I am a simple chatbot made with python.",
    "who made you": "A decodelabs intern made me using python.",
    "what is ai": "AI stands for Artificial Intelligence. It means making computers think and make decisions like humans.",
    "what is decodelabs": "DecodeLabs is a platform where interns learn AI by building real projects.",
    "what is project 1": "Project 1 is about making a rule based chatbot using if else logic and loops.",
    "what is ipo": "IPO means Input Process Output. First we take input, then process it, then show the output.",
    "help":"Try: hello, hi, how are you, what is ai\nwhat is ipo, who made you, bye. Type exit to quit.",
}

# default reply if bot doesnt understand
default_reply = "Sorry i dont understand that. Type help to see what i can do."

print("=" * 50)
print("  DecoBot v  |  DecodeLabs AI Training  2026")
print("=" * 50)
print("  Type 'help' to see commands | 'exit' to quit")
print("=" * 50)

# main loop - keeps running until user types exit
while True:

    # take input from user
    user_input = input("\nYou: ")

    # make it lowercase first
    user_input = user_input.lower()
    # then remove extra spaces
    user_input = user_input.strip()

    # check if user wants to quit
    if user_input == "exit" or user_input == "quit":
        print("DecoBot: Goodbye! Hope i was helpful.")
        break

    # check greetings using if else
    if user_input == "hello":
        print("DecoBot: " + responses["hello"])

    elif user_input == "hi":
        print("DecoBot: " + responses["hi"])

    elif user_input == "hey":
        print("DecoBot: " + responses["hey"])

    # check farewells
    elif user_input == "bye" or user_input == "goodbye":
        print("DecoBot: " + responses["bye"])

    # for everything else use dictionary lookup
    else:
        reply = responses.get(user_input, default_reply)
        print("DecoBot: " + reply)
```
